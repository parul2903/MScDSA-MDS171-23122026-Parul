```python
# Create a Python Class for managing student details and marks.
# Define the class and implement the methods of the student class in a menu-driven program 
# for different types of users

# STUDENT DETAILS CLASS AND METHODS 

import random

class student_details:

    def __init__(self):
        self.students = {}

    def add_details(self, name, register_number, email, sclass):

        self.students[register_number] = {
        "Name": name,
        "email":email,
        "sclass": sclass
        }

        print(self.students)

    def add_marks(self, regno, subject, marks):
        self.students[regno] = marks
    
    def search_details(self, register_number):     
        # on the basis of register number

        flag = False

        for key in self.students.keys():
            if key == register_number:
                print("The student details exist")
                print(self.students[key])
                flag = True

        if flag==False:
            print("The student details does not exist")

        
    def display_details(self, details):
        print(self.students[details])
    

    def update_details(self, regno, key, value):
        self.students[regno][key] = value
        print(self.students[regno])


# MENU-DRIVEN PROGRAM

print("Welcome to Christ University!")
print("----"*50)
print("This is the application to check student details and marks details of Christites.")
print("----"*50)

student = student_details()

while True:

    choice = int(input("MAKE A CHOICE: "))

    if choice == 1:
        name = input("Name: ")
        register_number = input('Register Number: ')  
        email = input('Email: ')
        sclass = input("Enter your class: ")

        student.add_details(name,register_number,email,sclass)
        print("the student details are succesfully added.")
        print("----"*50)

    elif choice == 2:
        regno = input("Enter the register number: ")
        subject = input("subject")
        marks = input("Marks")

        student.add_marks(regno,subject,marks)
        print("----"*50)
        print("Your marks are successfully added")

    elif choice == 3:
        search = input("Enter the register number: ")

        student.search_details(search)
        print("----"*50)

    elif choice == 4:
        regno = input("Enter the register number of students whose details you want to display: ")

        student.display_details(regno)
        print("----"*50)
        

    elif choice == 5:
        regno = input("Enter the register number of the student for which you want to update the detail: ")
        key = input("Enter the detail type you want to edit")
        value = input("Enter the detail")

        student.update_details(regno, key, value)
        print("Your details are successfully updated.")
        print("----"*50)

    else:
        break
```
